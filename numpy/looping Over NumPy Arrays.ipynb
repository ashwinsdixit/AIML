```python
import numpy as np
```

```python
mass=[12,20,25]
```

```python
mass_array = np.array(mass)
```

```python
for mass in mass_array:
    weight= mass * 9.8
    print(weight)
```

**<center>Looping over a 2D numpy array</center>**

```python
# create 2D numpy array using list of lists
massvolume = [[12,20,25],[36,18,10]]
```

```python
massvolume_array = np.array(massvolume)
```

```python
for value in massvolume_array:
    print (value)
```

```python
# Use nditer function to loop

for value in np.nditer(massvolume_array):
    print(value)
```

```python
#Take 1st row only
for value in np.nditer(massvolume_array[0]):
    weight = value * 9.8
    print(weight)
```
