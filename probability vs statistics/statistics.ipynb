**Coin toss example**

```python
import random

def coin_trail():
    heads = 0   

    for i in range(100):
        if random.random() <= 0.5:
            heads += 1
    return heads


def simulate(n):
    random.seed(123)
    trials = []
    for i in range(n):
        trials.append(coin_trail())
    return(sum(trials)/n)
```

```python
print("Average number of heads in 100 flips:", simulate(1000))
```
