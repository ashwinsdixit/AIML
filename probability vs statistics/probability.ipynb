**Deck of cards** Probability of drawing an ace from 52 card deck

```python
cards= 52

aces = 4

ace_probability = aces / cards

print(round(ace_probability, 2))
```

```python
ace_probability_percentage = ace_probability * 100
print(str(round(ace_probability_percentage, 2)) + "%")
```

**Prob of drawing a heart or a face card**

```python
def event_prob(event_outcomes, sample_space):
    probability = event_outcomes / sample_space
    return probability

cards = 52

hearts =13
hearts_probability = event_prob(hearts, cards)

face_cards = 12
face_cards_probability = event_prob(face_cards, cards)

print("Probability of drawing a heart:", round(hearts_probability, 2))
print("Probability of drawing a face card:", round(face_cards_probability, 2))
```
