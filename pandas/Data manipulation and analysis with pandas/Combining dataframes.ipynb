```python
import pandas as pd
```

```python
df_first_name_1 = pd.DataFrame({'Student_ID': ['S1234', 'S4321'], 'First Name': ['Daniel', 'Alfredo']})
df_first_name_2 = pd.DataFrame({'Student_ID': ['S3333', 'S4444'], 'First Name': ['Getrude', 'Ying']})   

df_last_names = pd.DataFrame({'Student_ID': ['S4321', 'S3333', 'S4444', 'S5678'], 'Last Name': ['Smith', 'Gupta', 'Minard', 'Gonzales']})
```

```python
df_first_name_1
```

```python
df_first_name_2
```

```python
df_last_names
```

**Union**<br> Specify axis='columns' to concatenate along the columns axis

```python
df_first_names = pd.concat([df_first_name_1, df_first_name_2]).reset_index(drop=True)
df_first_names
```

**Merge dataframe**

```python
#Inner join

pd.merge(df_first_names, df_last_names, on='Student_ID', how='inner')
```

```python
pd.merge(df_first_names, df_last_names, on='Student_ID', how='outer')
```

```python
pd.merge(df_first_names, df_last_names, on='Student_ID', how='left')
```

```python
pd.merge(df_first_names, df_last_names, on='Student_ID', how='right')
```
