**Create DataFrame**

```python
import pandas as pd
```

```python
rows = [
    ['Paul', 32, 'Software Engineer'],
    ['Sarah', 28, 'Data Scientist'],
    ['Stanley',9 , 'Dog']
    ]

pd.DataFrame(rows)
```

**Add columns and row labels with index and columns arguments**

```python
pd.DataFrame(rows, 
             columns=['Name', 'Age', 'Occupation'],
             index=['row1', 'row2', 'row3'])
```

**Using a dictionary to specify the column labels**

```python
family_dict = {'Name': ['Paul', 'Sarah', 'Stanley'],
               'Age': [32, 28, 9],
               'Occupation': ['Software Engineer', 'Data Scientist', 'Dog']}
pd.DataFrame(family_dict)
```

```python
df = pd.DataFrame(family_dict)
df.columns = ['days since getting a full night\'s sleep', 'Name','dream job']
df.index=['a', 'b', 'c']

df
```
