**map()**<br>
In Pandas, the map method is a primary tool for transforming data in a Series (a single column)

```python
import pandas as pd
import numpy as np
from pathlib import Path

# Load the data
data_path = ('employee_attrition.csv')  
original_df = pd.read_csv(data_path)
print(type(original_df))
```

```python
df = original_df.copy()

new_gender = df.Gender.map(lambda g: 'f' if g == 'Female' else 'm')

df.Gender = new_gender

df.head()
```

```python
df = original_df.copy()

lifer_col = df.TotalWorkingYears == df.YearsAtCompany
df['Lifer'] = lifer_col
df.head()
```

```python
df = original_df.copy()

df['unfair_compensation'] = df.DailyRate * df.JobLevel

df
```
