```python
import pandas as pd
import numpy as np
from pathlib import Path

df = pd.read_csv('employee_attrition.csv')
print(type(df))
df.head()
```

Selecting Data from DataFrame

```python
age = df['Age']
print(type(age))
age.head()
```

```python
sub_df = df[['Age', 'Attrition', 'DailyRate']]
print(type(sub_df))
sub_df.head()
```

```python
year_data = df[[col for col in df.columns if col.startswith('Year')]]
year_data.head()
```

**Selecting data with loc**

```python
row = df.loc[12:15]

print(type(row))
row
```

```python
age_of_emp = df.loc[12:15, 'Age']
print(type(age_of_emp))
age_of_emp
```
