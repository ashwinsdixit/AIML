```python
import pandas as pd
```

**Method1 : Convert dictionary to pandas df**

```python
 my_dict = {
    'college':['Columbia','Stanford','Harvard','IIT'],
    'accepted':[True,False,True,False]
 }

 #Convert dict to dataframe
 my_df = pd.DataFrame(my_dict)
 my_df
```

**Method 2 : Create a DataFrame from list of lists**

```python
list_of_lists = [['Columbia',True],['Stanford',False],['IIT',False]]

my_df = pd.DataFrame(list_of_lists , columns= ['college','accepted'])
my_df
```
