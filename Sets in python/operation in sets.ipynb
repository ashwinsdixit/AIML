Let's find the length of a set using `len()`

```python
my_set = {9, 3, 2, 4, 5}
len(my_set)
```

Let's perform some mathematical operations on sets using in-built functions

```python
sum(my_set)
```

```python
min(my_set)
```

```python
max(my_set)
```

We can sort a set in ascending or descending order

```python
sorted(my_set)  # returns a list as sets do not inherently have an order
```

```python
set(sorted(my_set))
```

```python
sorted(my_set, reverse = True)
```

```python
set(sorted(my_set, reverse = True))  # sets *appear* to be in ascending order no matter what; sorting a set means nothing
```

Let's check for membership of elements in sets

```python
s1 = {1, 2, 3, 4, 5}
s2 = {1, 2, 3, 4, 8}
```

```python
2 in s1
```

```python
8 in s2
```

```python
2 not in s1
```

Sets are mutable and therefore methods which modify sets act inplace, i.e., the modified set does not need to be reassigned to a variable for the changes to take effect

Here are some set methods

```python
s1.add(22)
s1
```

```python
s1.pop()  # removes and returns a random item from the set; sets do not have an order
```

```python
s1
```

```python
s1.remove(5)
s1
```

```python
s1.discard(3)
s1
```
