Let's create a set with values

```python
my_set = {1, 2, 3, 4}
my_set
```

```python
type(my_set)
```

```python
my_other_set = {9, 2, 7, 4, 0, 1}
my_other_set  # note that the output may seem ordered, but this isn't guaranteed by Python
```

```python
set(range(1, 10))
```

```python
set([1, 1, 2, 2, 2, 8, 8, 8, 3, 4])  # sets remove duplicate values; this is a good way to get uniques from sequences
```

```python
set('Hello world!')
```
